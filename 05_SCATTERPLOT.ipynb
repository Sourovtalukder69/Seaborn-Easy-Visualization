# Scatterplot

```python
# Bivariate analysis
# Scatter plot 
# `scatterplot` parameter use for create scatterplot in seaborn 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var = sns.load_dataset("penguins")
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex")
plt.show()

```

```python

# for working with small dataset use `head (how much you need)`
# for change style use `style=`
# for changing style use `size` and `sizes` parameters
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var = sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", style="sex", size="sex", sizes=(100, 40))
plt.show()


```

```python

# for change color  use `palette`
# `alpha` parameter use for visibility 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

var = sns.load_dataset("penguins")
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex",palette="Accent", alpha=0.5)
plt.show()

```

```python

# `marker changing`
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins").head(20)
mrks = {"Male":"*", "Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var,hue="sex",size= "sex", sizes=(30, 60), markers=mrks)
plt.show()
```
