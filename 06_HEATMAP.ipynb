# Heatmap

```python
# Hitmap : work by 2 dimentional data 
# `heatmap` function use to create heatmap
# `annot`=True parameter use to show values all box
# now we will create a simple heatmap of range 1 to 10  of 20 datas creating by numpy and then create heatplot 4 column 5 rows 

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
var = np.linspace(1, 10, 20).reshape(4, 5) # means give 20  value in range(1-10)
print(var)
sns.heatmap(var, annot=True)
plt.show()

```

```python

# Now we will create a heatmap from online data from github
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
data = sns.load_dataset("anagrams").head(10)
x = data.drop(columns=["attnr"], axis=1)   # for delet multiple use colums=["A","B"]
sns.heatmap(x)
plt.show()

```

```python

# heatmap minumu and maximum setting 
# `vmin=`, `vmax`
# `cmap` use for change entire color 

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
data = sns.load_dataset("anagrams").head(10)
x = data.drop(columns=["attnr"], axis=1)   # for delet multiple use colums=["A","B"]
sns.heatmap(x, vmin=0, vmax=12, cmap='Accent', annot=True)
plt.show()
```

```python
# crating an hetplot by my own value
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
var = np.linspace(1, 10, 10).reshape(2, 5)
ar = np.array([["Sourov", "Abir", "Jeny", "Any"," Aria"], ["age=20", "age=13", "age=16", "age=22", "age=2.5"]])
sns.heatmap(var, vmin=1, vmax=10, cmap='Accent', annot=ar, fmt="s")
plt.show()

```

```python

# `annot_kws` parameter use for changing font sizes and colors of data  
# `linewidth` parameter use for  modifying(gap) hetmap boxes ()
# `linelocor` parameter  use to change boxlines color

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
var = np.linspace(1, 10, 10).reshape(2, 5)
size_clr = {"fontsize":20, "color":"b"}

sns.heatmap(var, annot = True, annot_kws=size_clr, linewidths=10, linecolor="yellow")
plt.show()

# `cbar=False` use for rmv colorbar
# `xticklabels=False` use for remove all labels along x axis
# `yticklabels=False` use for remove all labels along y axis
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
var = np.linspace(1, 10, 10).reshape(2, 5)
size_clr = {"fontsize":20, "color":"b"}

sns.heatmap(var, annot = True, annot_kws=size_clr, linewidths=10, linecolor="yellow", cbar=False,xticklabels=False, yticklabels=False)
plt.show()

```

```python

# creating own label

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
var = np.linspace(1, 10, 10).reshape(2, 5)
size_clr = {"fontsize":20, "color":"b"}

v = sns.heatmap(var, annot = True, annot_kws=size_clr, linewidths=10, cbar=False,xticklabels=False, yticklabels=False)
plt.show()
v.set(xlabel="Python", ylabel="Sourov")
sns.set(font_scale=1)
plt.show()
```
