# Histogram

```python
# Univariate analysis -- Frequency count
#  creating a seaborn histogram by `displot` method
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins")
print(var.to_string())
sns.displot(var["bill_length_mm"])
plt.show()

```

```python

# for organize histogram more perfectly we can use `bins`
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins")
sns.displot(var["flipper_length_mm"],bins=[170, 180, 190, 200, 210, 220, 230, 240] )
plt.show()


```

```python

# for showing Kernal dencity estimator use `kde=True` parameter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins")
sns.displot(var["flipper_length_mm"],bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True)
plt.show()

```

```python
# for showing kde along x axis use `rug=True` parameter
# color parameter use for coloring
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins")
sns.displot(var["flipper_length_mm"],bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True, rug=True, color="green")
plt.show()
```

```python
# `log_scale parameter use for log according graph`

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("penguins")
sns.displot(var["bill_length_mm"],kde=True, rug=True, log_scale=True)
plt.show()

```
