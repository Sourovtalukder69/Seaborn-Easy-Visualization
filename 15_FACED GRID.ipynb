# Faced Grid

```python
# Facet Grid() function is basically use for plot multiple graph for different data

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
fg = sns.FacetGrid(x, col="sex", hue="day", )
fg.map(plt.scatter, "total_bill", "tip")
plt.legend()
plt.show()

```
