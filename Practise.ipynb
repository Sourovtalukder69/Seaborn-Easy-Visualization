```python
# import seaborn as sns
# seaborn is a visualization library on top of matplotlib and closely integreted with python pandas libraty
# creating a lineplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
x_var = [1, 6, 4, 8, 3]
y_var = [10, 60, 43, 67, 20]
sns.lineplot(x=x_var, y = y_var)
plt.show()
```

```python
# From dataframe
df = pd.DataFrame(
    {
        'math':[1, 6, 4, 8, 3],
        'english':[10, 60, 43, 67, 20]
    }
)
df
```

```python
sns.lineplot(data=df, x='math', y='english')
plt.show()
```

```python
pg = sns.load_dataset('penguins')
pg
```

```python
plt.style.use('fivethirtyeight')
sns.lineplot(data=pg, x='bill_length_mm', y='bill_depth_mm', hue='sex', markers=["o",">"], dashes=False)
plt.title('My_seaborn', fontdict={'size':20, 'c':'r'})
plt.show()
```

```python

# `order` parameter use to change graph along x
# `ci` parameter use to change the sign size above bars (err)
# `errcolor` parameter use for change errcolor
# `errwidth` use for change err width
# `capsize` parameter use for cap show
plt.style.use('default')
# orient use for horizonatal barplot
sns.barplot(data=pg, x='island', y='bill_length_mm',hue='sex', saturation=1 )
plt.show()
```
