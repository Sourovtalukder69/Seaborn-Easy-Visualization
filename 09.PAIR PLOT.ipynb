# Pair Plot

```python
# Pair plot : It is plot with respect to numeric dataset of a DataFrame or csv file
# Use for comparism use when machine learning algorithms use 

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips").head(10)
print(x)
sns.pairplot(x, hue="sex")
plt.show()

```

```python

# creating pairplot for some particular data
# `hue_order` use for orient for my choice
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips").head(10)
print(x)
sns.pairplot(x,vars=["total_bill", "tip"], hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python
# `x_bars=[]` , `y_bars=[]` use for plot my choice datas
# `kind` parameter use for change graph style under kind u can use "scatter", "kde", "hist", "reg"
# `diag_kind` use for fix types plot
# `markers` paramter use for change markers.
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips").head(10)
sns.pairplot(x, hue="sex", hue_order=["Female", "Male"], x_vars=["total_bill", "tip"], kind="kde")
plt.show()

#############
# sns.relplot(data=df, x='km_driven', y='selling_price', col='brand', col_wrap=4)
```
