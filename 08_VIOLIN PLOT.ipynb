# Violinplot

```python
# Violin plot : A violin plot plays a similar role as a box and whisker plot. It shows the distribution if quantitative data across several levels of one (or more) cateragoriacal variables such that those distribution can be compared. 

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")

sns.violinplot(x="day", y="total_bill", data=x, hue="time", color="r")
plt.show()

```

```python
# to increase line `linewidth` parameter use.
# for change colors use `pallete` parameter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=x, hue="time", linewidth=2, palette="Dark2")
plt.show()
```

```python
# `order` parameter use for change order along x axis
# `saturation` parameter use for brightness.range(0-1)

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips")
sns.violinplot(x="time", y="total_bill", data=x, linewidth=2, palette="Dark2", order=["Dinner", "Lunch"], saturation=0.5,  )
plt.show()
```

```python
# `split` parameter use for change the plot according to a specific term like Male ,Female o
# `scale` parameter use for chage with respect to "width", "count", area, 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=x,hue="sex", split=True,  scale="count")
plt.show()

```

```python

# to create horizontal plot change x and y value
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips")
sns.violinplot(x="total_bill", y="day", data=x,hue="sex", split=True)
plt.show()
```

```python
# `inner` parameter use for modify violin like "box", "quartile", "point", "stick"
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=x,hue="sex",scale="count", inner="quart")
plt.show()

```
