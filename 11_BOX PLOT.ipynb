# Box Plot

```python
# Box plot : Mainly use Matplotlib plot 
# Now we will create a boxplot from github data

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips")
sns.set(style="whitegrid")
sns.boxplot(x="day", y="total_bill", data=x)
plt.show()
```

```python

# use `hue` parameter for show with respect to specific term
# `order` parameter use for change its order like friday saturday sunday
# `showmeans=True` use for showing mean 
# `meanprops` use for change mean showing symboll modify
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips")
sns.set(style="whitegrid")
sns.boxplot(x="day", y="total_bill", data=x, hue="time",showmeans=True, meanprops={"marker":"+", "color":"r"}, linewidth=3, saturation=1, palette="plasma")
plt.show()

```

```python

# `orient` parameter use for showing grap horizontally. But have to change data because it required numeric data
# but dont need to use orient parameter because u can easily show grap horizontally by exchange x, y values
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x = sns.load_dataset("tips")
sns.set(style="whitegrid")
sns.boxplot(x="total_bill", y="day", data=x)
plt.show()

```
