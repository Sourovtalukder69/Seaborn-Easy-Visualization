```python
# Now we will create a lineplot 
# import matplotlib.pyplot as plt
# import seaborn as sns

# data = sns.load_dataset(" ")
# sns.lineplot(x=" ", data=data)
# plt.show()

import matplotlib.pyplot as plt
import seaborn as sns

var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

# matplotlib can easily plot with respect to two dataset 
plt.plot(var, var_1)
plt.show()

# But seaborn can't plot  because it can plot DataFrame 
sns.lineplot(x="var", y="var_1")


```

```python
# Creating a lineplot in seaborn 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

x_1 = pd.DataFrame({"var":var, "var_1":var_1})
sns.lineplot(x="var",y="var_1", data=x_1)
plt.show()
```

```python

# creating a line plot from github seaborn data
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins")
print(y_1)
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=y_1)
plt.show()
```

```python
# create graph with respect to popular dataset by : `hue=` parameter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex")  
plt.show()
```

```python
# you can change line size and style by `size=` parameter and `style=` parameter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", size=10, style='sex')  
plt.show()
```

```python
# for change color we use `palette=` parameter 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, palette="Accent")  
plt.show()
```

```python
# for create  marker use `marker=` parameter
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", style='sex', markers=["o", ">"])  
plt.show()
```

```python
# for working with low data use head() function
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins").head(20)
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", style='sex', markers=["o", ">"])  
plt.show()
```

```python
# removing dashesh
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins").head(20)
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", style='sex', markers=["o", ">"], dashes=False)  
plt.show()
```

```python
# also use some matplotlib functions like legend(), grid(), title() for more visualize
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
y_1 = sns.load_dataset("penguins").head(20)
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", style='sex', markers=["o", ">"], dashes=False, legend=False) 
plt.grid()
plt.title("My Seaborn")  # also can use `fontdict=`{'family':'serif', 'color':'red', 'size':'20'}
plt.show()
```

```python
# create a nice showing
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

info = sns.load_dataset("penguins")
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=info, hue="sex", style="sex", dashes=False,markers=["o",">"] )
font= {'family':'serif', 'color':'red', 'size':20}
plt.title("The Seaborn graph", fontdict=font)
plt.grid()
plt.show()
```
