# Strip plot

```python
# Strip plot : A strip plot is a verry simple to understand. It is basically a scatter plot that differentiates different categiries. strip plots are considered a good alternative to a box plot or a violin plot

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
sns.stripplot(x = "day", y="total_bill", data=x, hue="sex")
plt.show()

```

```python

# `palette` parameter use for color change
# `linwidth` parameter use for change line width
# `edgecolor` parameter use for change edge color
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
sns.stripplot(x = "day", y="total_bill", data=x, hue="sex", linewidth=3, edgecolor="green")
plt.show()
```

```python

# `jitter` parameter use for displacement change . its separates each bubble(like scatterplot) 
# `size` parameter use for change round size  
# `alpha` parameter use for brightness
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x = sns.load_dataset("tips")
m = {"Male":"o","Female":"*"} # use marker=m inside stripplot
sns.stripplot(x = "day", y="total_bill", data=x, hue="sex",jitter=1, size=20, alpha=0.5)
plt.show()

```
