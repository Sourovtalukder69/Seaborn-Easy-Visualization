# Barplot

```python
# Bivariate analysis
# Barplot : for creating barplot in seaborn need to parameter(x, y).
# Barplot consist by comparism of two parameter

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
# sns.barplot(x=var.island, y = var.bill_length_mm)  # it will aslo work
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.show()

```

```python

# for showing the barplot separately for specific condition like here we will separate with respect to sex=Male, Female by use `hue` parameter
# `hue` parameters order also change by use `hue_order=`
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python

# `order` parameter use to change graph along x
# `ci` parameter use to change the sign size above bars (err)
# `errcolor` parameter use for change errcolor
# `errwidth` use for change err width
# `capsize` parameter use for cap show
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, ci=100, errcolor="blue", errwidth=10)
plt.show()

```

```python
# `orient` parameter use for horizontal barplot but required x and y must be numerical
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
sns.barplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", orient='h')
plt.show()

# for coloring use `color ` parameter
# alos use `palette`
# `saturation` parameter use for brightness aso use `alpha` parameter
# `dodge` parameter use for ploting bar in bar
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order=order_1, palette='icefire', saturation=10, errcolor='green')
plt.show()
```
