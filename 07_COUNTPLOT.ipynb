# Countplot
### What is the difference between countplot and barplot ?
- 1.Countplot plots the count of the number of records by category                        
- 2.Barplot plots a value or metric of each category (by default barplot plots the mean of a variable, by catefory)

```python
# Creating countplot using `countplot` parameter from data 
# It show how much number
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
sns.countplot(x="sex", data=var)
plt.show()
```

```python

# creating bar plot from same data 
# it show mean value 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
sns.barplot(x="sex",y="size", data=var)
plt.show()
```

```python


# measuring specificly
# `saturation` parameter use to bright its value range(0-1)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
sns.countplot(x="sex", data=var, hue="smoker", palette="Accent", saturation=0.5,)
plt.show()
```
