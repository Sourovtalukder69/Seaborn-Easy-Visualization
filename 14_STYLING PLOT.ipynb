# Styling Plot

```python
# Styling plots in seaborn 
#   Seaborn figure styles
#   Removing Axes Spines
#   scale and context

# for changing grid 
import matplotlib.pyplot  as plt
import seaborn as sns
import pandas as pd

x = sns.load_dataset("tips")
sns.barplot(x ="day",y="total_bill", data=x, errcolor="b")
sns.set_style("darkgrid")  # "whitegrid", "darkgrid", white, dark 
plt.grid(axis='y')                
plt.show()

```

```python

# `despine` parameter use for removing axis up and right
import matplotlib.pyplot  as plt
import seaborn as sns
import pandas as pd

x = sns.load_dataset("tips")
sns.barplot(x ="day",y="total_bill", data=x,)
sns.despine()             
plt.show()

```

```python

# To change figure size 
import matplotlib.pyplot  as plt
import seaborn as sns
import pandas as pd

x = sns.load_dataset("tips")
sns.barplot(x ="day",y="total_bill", data=x,)
plt.figure(figsize=(3, 10))      
plt.show()

```

```python

# `set_contex` function 
import matplotlib.pyplot  as plt
import seaborn as sns
import pandas as pd

x = sns.load_dataset("tips")
sns.barplot(x ="day",y="total_bill", data=x, palette="cool", saturation=1) 
sns.set_context("paper", font_scale=10)    
plt.show()
```
